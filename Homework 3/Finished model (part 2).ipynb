```python
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
import pandas as pd
# import torch
import math
import statistics
from sklearn.model_selection import train_test_split
import random
```

```python
def xy_splitter(leaf_outline):
    x_list=[]
    y_list=[]
    leaf_outline=leaf_outline.split("|")
    for pixels in range(len(leaf_outline)):
        leaf_outline[pixels]=leaf_outline[pixels].split(",")
        leaf_outline[pixels][0]=int(leaf_outline[pixels][0])
        leaf_outline[pixels][1]=int(leaf_outline[pixels][1])
        x_list.append(leaf_outline[pixels][0])
        y_list.append(leaf_outline[pixels][1])

    return leaf_outline, x_list, y_list
```

```python
def min0(leafXorY):
    x = [pixs[0] for pixs in leafXorY]
    y = [pixs[1] for pixs in leafXorY]
    while min(x) !=0:
        leafXorY = [[pixs[0]-1,pixs[1]] for pixs in leafXorY]
        x = [pixs[0] for pixs in leafXorY]
    while min(y) !=0:
        leafXorY = [[pixs[0],pixs[1]-1] for pixs in leafXorY]
        y = [pixs[1] for pixs in leafXorY]
    return leafXorY
```

```python
def scale(leaf,width,height):
    scaled_leaf = [(pixs[0]/width, pixs[1]/height) for pixs in leaf]
    return scaled_leaf
```

```python
def SDEV(leaf):
    x = [pixs[0] for pixs in leaf]
    y = [pixs[1] for pixs in leaf]
    meanx = statistics.median(x)#mean
    meany = statistics.median(y)#mean
    SDEV_leaf=[[pixs[0]-meanx, pixs[1]-meany] for pixs in leaf]
    return SDEV_leaf
```

```python
def minDist(unkown_leaf, leaf_list):
    smallest_dist_array=[]
    smallest_dist_array=[]
    for _ in range(len(leaf_list)):
        xy_array=[]
        for xy in range(2):
            list1=[pixs[xy] for pixs in unkown_leaf]
            list2=[pixs[xy] for pixs in leaf_list.iloc[_]]
            larger_list=max(len(list1), len(list2))
            
            if larger_list != len(unkown_leaf):
                list_=list1
                list1=list2
                list2=list_

            smallest_dist=math.dist(list1[0:len(list2)],list2)
            for entries in range(len(list1)):
                list1_0=list1[0]
                list1.pop(0)
                list1.append(list1_0)
                distance=math.dist(list1[0:len(list2)],list2)
                if smallest_dist > distance:
                    smallest_dist = distance
            xy_array.append(smallest_dist)
        smallest_dist_array.append(xy_array[0]+xy_array[1])
    smallest_dist_index=smallest_dist_array.index(min(smallest_dist_array))
    return smallest_dist_index
```

```python
data = pd.read_csv("leaf outlines.csv")
# print(data.describe())

# print(data.keys())
data["Whole_leaf"] = data["Leaf_route_1"] + "|" + data["Leaf_route_2"] + "|" + data["Leaf_route_2"]

genus_species = data[['Genus', 'species']].astype(str)
train_data, test_data = train_test_split(genus_species,test_size=0.2,stratify=genus_species,random_state=42)
# print(test_data)

train_data = data[data[["Genus","species"]].astype(str).isin(train_data).all(axis=1)]
test_data = data[data[["Genus","species"]].astype(str).isin(test_data).all(axis=1)]

# print(train_data.keys())

# # print(train_data["Whole_leaf"].iloc[0])
for _ in range(len(train_data["species"])):
    new_value,x,y=xy_splitter(train_data["Whole_leaf"].iloc[_])
    train_data["Whole_leaf"].iloc[_]=new_value
    train_data["Whole_leaf"].iloc[_]=min0(train_data["Whole_leaf"].iloc[_])
    train_data["Whole_leaf"].iloc[_]=scale(train_data["Whole_leaf"].iloc[_],train_data["Leaf width"].iloc[_],train_data["Leaf height"].iloc[_]) 
    train_data["Whole_leaf"].iloc[_]=SDEV(train_data["Whole_leaf"].iloc[_])
# print(train_data["Whole_leaf"].iloc[0])

for _ in range(len(test_data["species"])):
    new_value,x,y=xy_splitter(test_data["Whole_leaf"].iloc[_])
    test_data["Whole_leaf"].iloc[_]=new_value
    test_data["Whole_leaf"].iloc[_]=min0(test_data["Whole_leaf"].iloc[_])
    test_data["Whole_leaf"].iloc[_]=scale(test_data["Whole_leaf"].iloc[_],test_data["Leaf width"].iloc[_],test_data["Leaf height"].iloc[_]) 
    test_data["Whole_leaf"].iloc[_]=SDEV(test_data["Whole_leaf"].iloc[_])
```

```python
rand_data=[]
while len(rand_data) != len(range(len(test_data))):
    i=random.choice(range(len(test_data)))
    if i not in rand_data:
        rand_data.append(i)

accuracy=[0,0]
for _ in rand_data[0:12]:
    target_leaf_index=_
    target_leaf=test_data["Whole_leaf"].iloc[target_leaf_index]
    leaf_array=train_data["Whole_leaf"]
    suggested_individual=minDist(target_leaf,leaf_array)

    target=test_data[["Genus","species"]].astype(str).iloc[_]
    suggested=train_data[["Genus","species"]].astype(str).iloc[suggested_individual]
    target=target[0]+target[1]
    suggested=suggested[0]+suggested[1]
    
    accuracy[1]+=1
    if target==suggested:
        accuracy[0]+=1
    print("Score:",accuracy[0], "/", accuracy[1])
    print("Accuracy:", accuracy[0]/accuracy[1])
    print(test_data["Individual"].iloc[_])
    print(train_data["Individual"].iloc[suggested_individual])
```
